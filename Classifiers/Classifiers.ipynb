```python
import numpy as np
import pandas as pd
data =pd.read_excel('data.xlsx',sheet_name='data')
df=pd.DataFrame(data)
```

```python
df.describe()
```

```python
df.info()
```

```python
#Dropping Columns With NaN values.
df=pd.DataFrame(data)
df=df.dropna(axis=1)
df.head()
```

```python
#Dropping unwanted columns
df=df.drop(['appointmentId','inspectionStartTime','year','month'],axis=1)
df.head()
```

```python
#Converting data into useable data
df=df.replace({'No': 0, 'Yes': 1})
df.head()
```

```python
#Plotting a boxplot of odometer_reading vs rating_engineTransmission
import seaborn as sns
sns.boxplot(x=df['rating_engineTransmission'],y=df['odometer_reading'])
```

It can be clearly seen that there are some outliers present in the dataset. The outliers are the points where the odometer_reading>500000. The threshold 500000 is obtained by observation. These data points are removed.

```python
idx=df[df['odometer_reading']>500000].index
df.drop(idx,inplace=True)
df.head()
```

```python
#Normalizing odometer_reading values
df['odometer_reading']=df['odometer_reading']-np.sum(df['odometer_reading'])/len(df['odometer_reading'])
df['odometer_reading']=df['odometer_reading']/np.std(df['odometer_reading'])
df.head()
```

```python
df1=pd.DataFrame(df)
df1.head()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
encoder.fit(df1['rating_engineTransmission'])
df1['rating_engineTransmission']= encoder.transform(df1['rating_engineTransmission'])
print(df1['rating_engineTransmission'].value_counts())
```

```python
encoder1=LabelEncoder()
encoder1.fit(df1['fuel_type'])
df1['fuel_type']= encoder1.transform(df1['fuel_type'])
print(df1['fuel_type'].value_counts())
```

```python
encoder2=LabelEncoder()
encoder2.fit(df1['engineTransmission_engineBlowByBackCompression_cc_value_0'])
df1['engineTransmission_engineBlowByBackCompression_cc_value_0']= encoder2.transform(df1['engineTransmission_engineBlowByBackCompression_cc_value_0'])
print(df1['engineTransmission_engineBlowByBackCompression_cc_value_0'].value_counts())
```

```python
Y=df1['rating_engineTransmission']
X=df1[:25000]
X_test=df1[25000:]
X=X.drop(['rating_engineTransmission','odometer_reading'],axis=1)
X_test=X_test.drop(['rating_engineTransmission','odometer_reading'],axis=1)
X=X.to_numpy()
X_test=X_test.to_numpy()
y=Y[:25000]
y_test=Y[25000:]
# print(X,y.shape,y)
```

```python
from sklearn.ensemble import RandomForestClassifier
rand = RandomForestClassifier()
rand.fit(X,y)
pred = rand.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred)
```
